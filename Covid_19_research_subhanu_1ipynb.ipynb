# Example of how to change some values in table

```
def change_state_name(state):
    if state == 'Odisha':
        return 'Orissa'
    elif state == 'Telengana':
        return 'Telangana'
    return state
```

```
covid_train['State/UnionTerritory'] = covid_train.apply(lambda x: change_state_name(x['State/UnionTerritory']), axis=1)
```

```
last_date = covid_train.DateTime.max()
covid_train = covid_train[covid_train['DateTime']==last_date]
```

```
col =['State/UnionTerritory', 'Cured', 'Deaths', 'Confirmed']
covid_train = covid_train[col]
covid_train.sort_values('Confirmed', inplace=True, ascending=False)
covid_train.reset_index(drop=True, inplace=True)
covid_train.style.background_gradient(cmap='PuBu')
```

```
col =['State/UnionTerritory', 'Cured', 'Deaths', 'Confirmed']
covid_train = covid_train[col]
covid_train.sort_values('Confirmed', inplace=True, ascending=False)
covid_train.style.background_gradient(cmap='viridis')
```

```
r = requests.get(url='https://raw.githubusercontent.com/deldersveld/topojson/master/countries/india/india-states.json')
geojson = r.json()
```

```
india_map = px.choropleth(covid_train, color="Confirmed", 
                      locations="State/UnionTerritory", geojson=geojson, featureidkey="properties.NAME_1",
                    hover_data=['Cured','Deaths'],
                    color_continuous_scale=px.colors.sequential.Viridis,
                    title='India: Map view of total current cases statewise')
india_map.update_geos(fitbounds="locations", visible=True)
india_map.update_geos(projection_type="orthographic")
india_map.update_layout(height=600,margin={"r":0,"t":30,"l":0,"b":30})
india_map.show()
```
